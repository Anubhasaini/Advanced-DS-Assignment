```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.preprocessing import LabelEncoder

# ✅ Correct path
data = pd.read_csv('/kaggle/input/datasets/kemical/kickstarter-projects/ks-projects-201801.csv')

# Select simple columns
data = data[['goal', 'state', 'currency']].dropna()

# Convert categorical → numeric
le = LabelEncoder()
data['state'] = le.fit_transform(data['state'])
data['currency'] = le.fit_transform(data['currency'])

# Features & target
X = data[['goal', 'currency']]
y = data['state']

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Ridge model
ridge = Ridge()
ridge.fit(X_train, y_train)
ridge_pred = ridge.predict(X_test)

# Lasso model
lasso = Lasso()
lasso.fit(X_train, y_train)
lasso_pred = lasso.predict(X_test)

print("Ridge Predictions:", ridge_pred[:5])
print("Lasso Predictions:", lasso_pred[:5])
```
